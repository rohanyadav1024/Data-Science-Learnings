```python
import numpy as np
import pandas as pd
import random
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, n_features=5, n_clusters_per_class=1,n_classes=2, n_informative=5, n_redundant=0)
```

```python
df = pd.DataFrame(X, columns=['col1', 'col2','col3','col4','col4'])
df['target'] = y
df
```

```python
df.head()
```

```python
def sample_rows(df, perc):
    return df.sample(int(perc * df.shape[0]), replace="True")
```

```python
def sample_features(df, perc):
    # cols = random.sample(df.columns.tolist()[:-1],int(percent*(df.shape[1]-1)))
    cols = random.sample(df.columns.tolist()[:-1], int(perc * (df.shape[1]-1)))
    new_df = df[cols]
    new_df['target'] = df['target']
    return new_df
```

```python
def sample_combine(df, row_perc, col_perc):
    return sample_features(sample_rows(df, row_perc), col_perc)
```
