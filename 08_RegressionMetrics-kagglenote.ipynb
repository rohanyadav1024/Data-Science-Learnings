```python
df = pd.read_csv("/kaggle/input/heart-diseaseuci/heart.csv")
```

```python
df.head()
```

```python
df.sample(10)
# y = target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(columns=["target"]), df["target"], test_size=0.25, random_state=2)
```

```python
X_train.sample(5), Y_train.sample(5)
```

```python
lor = LogisticRegression()
dt = DecisionTreeClassifier()
```

```python
X_train.shape
```

```python
lor.fit(X_train, Y_train)
```

```python
dt.fit(X_train, Y_train)
```

```python
y1_pred = lor.predict(X_test)
y1_pred
```

```python
y2_pred = dt.predict(X_test)
y2_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, f1_score, recall_score
```

```python
a1 = accuracy_score(Y_test, y1_pred)
a2 = accuracy_score(Y_test, y2_pred)
```

```python
a1, a2
```

<h2>Logistic Regression</h2>

```python
c1 = confusion_matrix(Y_test, y1_pred)
# for logistic regression
c1
print("precision score", precision_score(Y_test, y1_pred))
print("recall score", recall_score(Y_test, y1_pred))
print("f1 score", f1_score(Y_test, y1_pred))
```

<h2>Decision Tree</h2>

```python
c2 = confusion_matrix(Y_test, y2_pred)
c2
# for decision tree
print("precision score", precision_score(Y_test, y2_pred))
print("recall score", recall_score(Y_test, y2_pred))
print("f1 score", f1_score(Y_test, y2_pred))
```
